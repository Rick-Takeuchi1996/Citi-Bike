# Part 3 - Data Cleaning & Wrangling

```python
citi.head()
```

```python
citi.isnull().sum()
```

```python
#making subset of missing birth year values
na_birth = citi[citi['birth_year'].isnull()==True]
```

```python
na_birth.shape
```

```python
#drop na values
citi_clean = citi[citi['birth_year'].isnull()== False]
```

```python
#check if missing values are removed
citi_clean.shape
```

```python
citi_clean.head()
```

```python
# dropping unnessesary columns
citi_clean = citi_clean.drop(columns = ['trip_id'])
```

```python
citi_clean = citi_clean.drop(columns = ['bike_id'])
```

```python
#changing weekday to day_of_week
citi_clean = citi_clean.rename(columns={'weekday':'day_of_week'})
```

```python
citi_clean.info()
```

```python
#making subset users born after 1924
citi_clean_1 = citi_clean.loc[citi_clean['birth_year'] > 1928]
```

```python
citi_clean.shape
```

```python
citi_clean_1.shape
```

```python
citi_clean_1.head()
```

```python
#review unique values of 7 days of week
citi_clean_1['day_of_week'].unique()
```

```python
#changing gender to string
citi_clean_1['gender'] = citi_clean_1['gender'].astype('str')
```

```python
citi_clean_1['gender'] = citi_clean_1['gender'].values.astype('str')
```

```python
#changing variables to names of gender
citi_clean_1['gender'] = citi_clean_1['gender'].replace([0,1,2],['Unknown','Male','Female'])
```

```python
citi_clean_1.describe()
```

```python
citi_clean_1.head()
```

```python
#change start time to start of ride
citi_clean_1['start_of_ride'] = pd.to_datetime(citi_clean_1['start_time'],infer_datetime_format=True)
citi_clean_1['end_of_ride'] = pd.to_datetime(citi_clean_1['end_time'],infer_datetime_format=True)
```
